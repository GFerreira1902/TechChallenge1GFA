# 01 â€“ AnÃ¡lise ExploratÃ³ria de Dados (EDA)

## Objetivo
Entender os dados antes de qualquer modelagem. A EDA responde perguntas como:
- Quais variÃ¡veis temos disponÃ­veis?
- HÃ¡ valores faltando?
- Como estÃ£o distribuÃ­das as variÃ¡veis?

## Dataset: Breast Cancer Wisconsin
Dados de tumores de mama com caracterÃ­sticas extraÃ­das de imagens de biÃ³psia.  
**Alvo:** `diagnosis` â€” M (Maligno) ou B (Benigno)

```python
# ImportaÃ§Ãµes necessÃ¡rias
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

# ConfiguraÃ§Ã£o visual
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print('Bibliotecas carregadas com sucesso!')
```

## 1. Download e Carregamento dos Dados

Usamos o `kagglehub` para baixar o dataset automaticamente.  
Na primeira execuÃ§Ã£o ele farÃ¡ o download; nas seguintes usa o cache local.

> **PrÃ©-requisito:** ter a Kaggle API key configurada (`~/.kaggle/kaggle.json`)

```python
# Baixa o dataset automaticamente do Kaggle
path = kagglehub.dataset_download('uciml/breast-cancer-wisconsin-data')

print('Dataset salvo em:', path)
print('Arquivos disponÃ­veis:', os.listdir(path))
```

```python
# Carrega o CSV principal
csv_path = os.path.join(path, 'data.csv')
df = pd.read_csv(csv_path)

print(f'Shape do dataset: {df.shape}')  # (linhas, colunas)
df.head()
```

## 2. VisÃ£o Geral dos Dados

```python
# Tipos de dados e valores nulos por coluna
df.info()
```

```python
# EstatÃ­sticas descritivas
# Mostra: mÃ©dia, desvio padrÃ£o, mÃ­nimo, mÃ¡ximo e quartis de cada coluna numÃ©rica
df.describe()
```

```python
# Verifica valores nulos por coluna
nulos = df.isnull().sum()
print('Valores nulos por coluna:')
print(nulos[nulos > 0] if nulos.any() else 'Nenhum valor nulo encontrado!')
```

## 3. DistribuiÃ§Ã£o da VariÃ¡vel Alvo

Verificamos se hÃ¡ **desbalanceamento de classes** â€” um problema comum em dados mÃ©dicos,  
onde casos positivos (malignos) sÃ£o muito menos frequentes que negativos (benignos).  

Se o desbalanceamento for muito grande, o modelo pode "aprender" a sempre prever a classe majoritÃ¡ria e ainda assim ter alta accuracy â€” o que seria enganoso.

```python
contagem = df['diagnosis'].value_counts()
print(contagem)

plt.figure(figsize=(6, 4))
sns.countplot(x='diagnosis', data=df, palette='Set2')
plt.title('DistribuiÃ§Ã£o do DiagnÃ³stico (M=Maligno, B=Benigno)')
plt.xlabel('DiagnÃ³stico')
plt.ylabel('Quantidade')
plt.tight_layout()
os.makedirs('../outputs/figures', exist_ok=True)
plt.savefig('../outputs/figures/distribuicao_diagnostico.png', dpi=150)
plt.show()
```

## 4. AnÃ¡lise de CorrelaÃ§Ã£o

O mapa de calor (heatmap) mostra como as variÃ¡veis se relacionam entre si:
- **PrÃ³ximo de 1:** quando uma sobe, a outra sobe junto
- **PrÃ³ximo de -1:** quando uma sobe, a outra desce
- **PrÃ³ximo de 0:** sem relaÃ§Ã£o

VariÃ¡veis muito correlacionadas entre si podem ser redundantes â€” isso nos ajuda a escolher quais manter no modelo.

```python
# Seleciona apenas colunas numÃ©ricas
numericas = df.select_dtypes(include=np.number)

plt.figure(figsize=(18, 14))
sns.heatmap(numericas.corr(), cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Mapa de CorrelaÃ§Ã£o â€“ Breast Cancer Wisconsin')
plt.tight_layout()
plt.savefig('../outputs/figures/heatmap_correlacao.png', dpi=150)
plt.show()
```

## 5. DistribuiÃ§Ã£o das Principais Features

Visualizamos como as principais variÃ¡veis se distribuem para casos Malignos vs Benignos.  
Se uma variÃ¡vel separa bem as duas cores, ela tem alto poder preditivo.

```python
features_relevantes = [
    'radius_mean', 'texture_mean', 'perimeter_mean',
    'area_mean', 'smoothness_mean', 'concavity_mean'
]

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for i, feature in enumerate(features_relevantes):
    sns.histplot(
        data=df, x=feature, hue='diagnosis',
        kde=True, ax=axes[i], palette='Set2'
    )
    axes[i].set_title(f'DistribuiÃ§Ã£o: {feature}')

plt.suptitle('DistribuiÃ§Ã£o das Features por DiagnÃ³stico', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('../outputs/figures/distribuicao_features.png', dpi=150)
plt.show()
```
